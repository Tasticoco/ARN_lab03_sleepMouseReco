## Concatenation of the two dataframes

```python
import pandas as pd

mice_eeg1 = pd.read_csv('./pw3_data/EEG_mouse_data_1.csv')
mice_eeg2 = pd.read_csv('./pw3_data/EEG_mouse_data_2.csv')

mice_eeg = pd.concat([mice_eeg1, mice_eeg2], axis=0)

mice_eeg
```

## Data normalization

```python
import sklearn.preprocessing as skp
scaler = skp.StandardScaler()

mice_eeg_normalized = scaler.fit_transform(mice_eeg.iloc[1:])
mice_eeg_normalized = pd.DataFrame(mice_eeg_normalized, columns=mice_eeg.columns)
mice_egg_normalized.insert(0, mice_eeg.iloc[0])
mice_eeg_normalized
```
